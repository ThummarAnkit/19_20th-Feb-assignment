# Q1. Explain GET and POST methods.

__Ans. :__<br>

__GET__ and __POST__ are __HTTP__ methods used in web development to send data from a client (such as a web browser) to a server.

### GET Method:
The __GET__ method is used to request data from the server. It is an __HTTP__ method used to retrieve data from a specified resource, such as a webpage, and is the default method used by web browsers when a user clicks on a link or enters a URL in the address bar.

### POST Method:
The __POST__ method is used to submit data to the server. It is an __HTTP__ method used to send data to a server to create or update a resource.

# Q2. Why is request used in Flask?

__Ans. :__<br>

In Flask, request is a built-in object that allows you to access incoming request data in a Flask application. It is a global object that represents the current HTTP request and contains all the data that was sent with the request, such as form data, query string parameters, and HTTP headers.

Here are some of the common uses of request in Flask:

1. Accessing form data: When a user submits a form in a Flask application, the form data is sent as part of the request. You can use request.form to access the submitted form data in your Flask view function.

2. Accessing query string parameters: Query string parameters are a way of passing data in the URL, for example http://example.com?param1=value1&param2=value2. You can use request.args to access these parameters in your Flask view function.

3. Accessing HTTP headers: HTTP headers are used to pass additional information about the request or response, such as the user agent or the content type. You can use request.headers to access the headers sent with the current request.

4. Accessing cookies: Cookies are small pieces of data that are stored on the user's computer and are sent with each request to the same domain. You can use request.cookies to access the cookies sent with the current request.

Overall, the request object is a powerful tool in Flask that allows you to access and manipulate incoming request data in your Flask application.

# Q3. Why is redirect() used in Flask?

__Ans. :__<br>

redirect() is a function in Flask that is used to redirect the user to a different URL. It is commonly used in web applications to redirect users after they have submitted a form or completed an action.

Here are some common reasons to use redirect() in a Flask application:

1. After form submission: When a user submits a form in a Flask application, you may want to redirect them to a different page instead of rendering the same page again. This is often done to prevent the user from accidentally resubmitting the form if they refresh the page.

2. After authentication: When a user logs in to a Flask application, you may want to redirect them to a different page that is only accessible to authenticated users.

3. After a successful action: After a user completes an action, such as creating a new account or making a purchase, you may want to redirect them to a "success" page to confirm that the action was completed successfully.

# Q4. What are templates in Flask? Why is the render_template() function used?

__Ans. :__<br>

In Flask, templates are files that are used to generate dynamic HTML pages. Templates are usually written in HTML, with placeholders for dynamic data that is inserted at runtime. Flask uses the Jinja2 template engine, which allows for advanced features such as template inheritance, macros, and loops.

The __render_template()__ function in Flask is used to render a template with the given context. The context is a dictionary that contains the data that will be inserted into the template. The __render_template()__ function takes two arguments: the name of the template file (relative to the __templates__ directory in your Flask project), and the context dictionary.
